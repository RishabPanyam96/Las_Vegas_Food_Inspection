## Checking yelp data

```python
# Checking the business.csv dataframe
business = pd.read_csv('business.csv')
print('Entire business data set shape:', business.shape)
business.head()
```

```python
# Since the scope of the project is for cities near the Las Vegas area, We will subset the data for businesses in Nevada

#******* Should we clean the data to include misspellings and close associations??? ************
business_NV = business[business['state'] == 'NV']
# (36312, 14)
# print('State of Nevada subset shape: ', business_NV.shape)
business_NV['city'].value_counts()

## Subsetting only Las Vegas
# (29361, 14)

business_LV = business_NV[business_NV['city'] == 'Las Vegas']
print('City of Las Vegas subset shape: ', business_LV.shape)
business_LV.head(5)
```
