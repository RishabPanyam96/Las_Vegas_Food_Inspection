```python
def get_levinshtein_distance_city(df, city_name, city_column_name):
    df['city_'] = df[city_column_name].apply(lambda x: x.lower().strip().title())
    column_name = city_name[:1] + '_levi_dist'
    df[column_name] = df.loc[:, 'city_'].apply(lambda x: editdistance.eval(x, city_name))
# function to get levensitein distance for each row of the dataframe  
```

```python
def clean_city_name(df, replace_with, city_column_name, cutoff):
    column_name = replace_with[:1] + '_levi_dist'
    df.loc[df[column_name] < cutoff, city_column_name] = replace_with
#get clean city name for each columns    
```

```python
list_remove = ['bar','cafe','grill','casino','restaurant','hotel','resort','kitchen','pizza','lounge','bakery','shop','market']
# list of words to remove
```

```python
def clean_text(x):
    temp = str(x)
    temp = temp.lower()
    tokens = word_tokenize(temp)
    words = [word for word in tokens if word.isalpha()]
    words = [w for w in words if not w in stop_words]
    words = [w for w in words if not w in list_remove]
    k= " ".join(words)
    return (k.strip()) 
#cleaning the name,removing punctuations,removing frequent words
```

```python
def get_LCS(x,y):
    s = SequenceMatcher(lambda x: x==' ',x,y)
    match = s.find_longest_match(0, len(x), 0, len(y)) 
    if (len(x) >= len(y)):
        min_len = len(y)
    else: min_len = len(x)    
    lcs = x[match.a: match.a + match.size]
    return [lcs,len(lcs),len(lcs)/(min_len+1)]
# fucntion to return LCS,len(LCS),normalized_LCS
```

```python
def get_nearest_facid_name(x,df):
    df_filter = df[df['latitude'].between(x['latitude'] -0.001,x['latitude']+0.001) & df['longitude'].between(x['longitude']-0.001,x['longitude']+0.001)]
    if (df_filter.shape[0] <=10) :
        df_filter = df[df.latitude.between(x['latitude'] -0.01,x['latitude']+0.01) & df.longitude.between(x['longitude']-0.01,x['longitude']+0.01)]
    df_filter['business_name'] = x['name']
    df_filter['business_address'] = x['address']
    if df_filter.empty:
        return None   
    df_filter['lev_dist_normalized'] = df_filter['clean_loc_name'].apply(lambda y:editdistance.eval(x['clean_name'],y)/(len(x['clean_name'])+len(y)+1))
    df_filter['LCS'] = df_filter.apply(lambda y:get_LCS(x['clean_name'],y['clean_loc_name'])[0],axis =1 ,result_type ='expand')
    df_filter['LCS_len'] = df_filter.apply(lambda y:get_LCS(x['clean_name'],y['clean_loc_name'])[1],axis =1 ,result_type ='expand')         
    df_filter['LCS_normal'] = df_filter.apply(lambda y:get_LCS(x['clean_name'],y['clean_loc_name'])[2],axis =1 ,result_type ='expand')
    #df_filter['LCS_normal'] = df_filter['clean_rest_name'].apply(lambda y : (lcs.llcs(x['clean_name'], y))/(len(x['clean_name'])+len(y)))
    df_filter = df_filter.sort_values(by=['LCS_normal'],ascending = False)
    ret = df_filter[df_filter.LCS_normal>0.5]
    if ret.empty:
        return None
    ret =ret['facility_id'].unique()
    return ret
# function to return fcaility_ids for each name in business_Df from establishment_df
```

```python

def get_nearest_dist(x,df):
    dist =1000000000000000000
    name=''
    facility_id=''
    df_filter = df[df.latitude.between(x.latitude -0.001,x.latitude+0.001) & df.longitude.between(x.longitude-0.001,x.longitude+0.001)]
    for i in df_filter.index:
        distance = (df['latitude'][i]-x.latitude)**2 + (df['longitude'][i]-x.longitude)**2
        if(distance<dist ):
            name_in = df['location_name'][i]
            address_in = df['address'][i]
            facility_id = df['location_name'][i]
            dist = distance
    return [name_in,facility_id,address_in,dist]   
```

```python

def get_nearest_name(x,df):
    lev_dist =1000000000000000000
    name=''
    facility_id=''
    df_filter = df[df.latitude.between(x.latitude -0.001,x.latitude+0.001) & df.longitude.between(x.longitude-0.0005,x.longitude+0.0005)]
    for i in df_filter.index:
        lev_dist_temp = editdistance.eval(x['name'].lower(),df['location_name'][i].lower())
        if(lev_dist_temp<lev_dist ):
            lev_dist = lev_dist_temp
            name_in = df['location_name'][i]
            address_in = df['address'][i]
            facility_id = df['location_name'][i]
            distance = (df['latitude'][i]-x.latitude)**2 + (df['longitude'][i]-x.longitude)**2
    if (lev_dist<20):        
         return [lev_dist,name_in,facility_id,address_in,distance]
    else:
        return [None,None,None,None,None]
```
