### Split the business rows with multiple permits

```python
multiple_ind = business_df.apply(lambda x: True if len(x['key']) > 13 else False, axis=1)
multiple_permit_count = len(multiple_ind[multiple_ind == True].index)

single_ind = [not x for x in multiple_ind]

multiple_ind_true = multiple_ind[multiple_ind == True].index
multiple_permit_count
```

```python
business_single_df = business_df.copy()[single_ind]

for row in list(multiple_ind_true):
    keys = business_df.loc[row, "key"]
    current_row = business_df.loc[row, :]
    for key in keys[2:len(keys)-2].split():
        copy_row = current_row.copy()
        copy_row["key"] = key.replace("'", "")
        business_single_df = business_single_df.append(copy_row, ignore_index = True)
            
#     business_df.drop(row, inplace = True)
    print("dropped", row)
```

```python
multiple_ind = business_single_df.apply(lambda x: True if len(x['key']) > 13 else False, axis=1)
multiple_permit_count = len(multiple_ind[multiple_ind == True].index)
multiple_permit_count
```

```python
from datetime import datetime, timedelta

time_diff = timedelta(days=30)

unique_inspections = business_ment_df.groupby(['facility_id']).agg({'date_current': pd.Series.nunique})
facility_multiple_inspections = unique_inspections[unique_inspections['date_current'] > 1].index
```

```python
#TODO: just apply the strptime function to the whole column

for facility in facility_multiple_inspections:
    facility_rows = business_ment_df[business_ment_df['facility_id'] == facility]
    inspection_dates = [datetime.strptime(x, '%Y-%m-%d %H:%M:%S') for x in 
                        sorted(facility_rows['date_current'].drop_duplicates().tolist())]

    for idx, date in enumerate(inspection_dates):
        if idx == len(inspection_dates) - 1:
            print("idx limit", False)
        else:
            if (date + time_diff) > inspection_dates[idx + 1]:
                print(date + time_diff, inspection_dates[idx + 1], True)
            else:
                print(False)
```
